## Feature_Engineering

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
train[train['DAYS_EMPLOYED']>0]#4438
```

```python
test[test['DAYS_EMPLOYED']>0]#1697 
```

train과 test에 DAYS_EMPLOYED가 양수인 부분을 0으로 바꿔주고 DAYS_EMPLOYED가 0이면서 occyp_type이 NAN 값인 부분을 무직자(missing_job)로 전환하고 나머지 결측값을 하나의 카테고리(unknown)로 전환하겠다.

```python
train['DAYS_EMPLOYED'] =train.apply(lambda x: 0 if (x.DAYS_EMPLOYED>0) else x.DAYS_EMPLOYED,axis=1)
test['DAYS_EMPLOYED'] =test.apply(lambda x: 0 if (x.DAYS_EMPLOYED>0) else x.DAYS_EMPLOYED,axis=1)
```

```python
train['occyp_type'] =train.apply(lambda x: 'missing_job' if (x.DAYS_EMPLOYED==0) else x.occyp_type,axis=1)
test['occyp_type'] =test.apply(lambda x: 'missing_job' if (x.DAYS_EMPLOYED==0) else x.occyp_type,axis=1)
```

```python
train.isnull().sum()
test.isnull().sum()
```

남아있는 결측치
- train : 3733개
- test : 1455개

```python
train= train.fillna('unknown')
test=test.fillna('unknown')
train.isnull().sum()
```

index열 지우기

```python
del train['index']
del test['index']
```

```python
train[['occyp_type']]
```

occyp_type에  결측값을 하나의 카테고리로 분류해서 전처리 as 'Missing_Job'

### 원핫인코딩

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
object_col = []
for col in train.columns:
    if train[col].dtype == 'object':
        object_col.append(col)
object_col
```

```python
encoding = OneHotEncoder()
encoding.fit(train.loc[:,object_col])


train_onehot = pd.DataFrame(encoding.transform(train.loc[:,object_col]).toarray(), 
             columns=encoding.get_feature_names(object_col))
train.drop(object_col, axis=1, inplace=True)
train = pd.concat([train, train_onehot], axis=1)
train


```

```python
test_onehot = pd.DataFrame(encoding.transform(test.loc[:,object_col]).toarray(), 
             columns=encoding.get_feature_names(object_col))
test.drop(object_col, axis=1, inplace=True)
test = pd.concat([test, test_onehot], axis=1)
test
```

DAYS_BIRTH는 나이로 바꿔주고, DAYS_EMPLOYED	

```python
train.to_csv('data/train_onehot.csv')
test.to_csv('data/test_onehot.csv')
```
