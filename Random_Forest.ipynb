### train_x,train_y, test_x, test_y 만들기

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('data/train_onehot.csv')
test=pd.read_csv('data/test_onehot.csv')
```

```python
train_x=train.drop('credit',axis=1)
train_y=train['credit']
```

```python
print(train_x.shape)
print(train_y.shape)
```

```python
test_x=test
```

```python
a=pd.concat([train,test],axis = 0)
b=a[-10000:]
test_y=b['credit']
```

```python
print(test_x.shape)
print(test_y.shape)
```

randomforest이용하기

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(train_x,train_y)
```

```python
train_proba=clf.predict_proba(train_x)
train_proba
```

```python
np.argmax(train_proba, axis=1)[:10]
```

```python
clf.predict_proba(test_x)
```

```python
sample_submission=pd.read_csv('data/sample_submission.csv')
sample_submission
```

```python
sample_submission.iloc[:,1:]=clf.predict_proba(test_x)
```
